```python
#importar databdes from Gbif
from pygbif import species as species 
from pygbif import occurrences as occ
import pygbif
pygbif.caching(True)
import pandas as pd
```

```python
#verify existance of Mycobacterium sp and Mycobacterium tuberculosis in db
splist = ['Mycobacterium sp','Mycobacterium tuberculosis']
keys = [species.name_backbone(x)['usageKey'] for x in splist]
keys
```

```python
# count with taxon key Gbif
out = occ.search(taxonKey = 3225140, limit=0)['count'] 
out
```

```python
#total count of taxon
outTotal = [occ.search(taxonKey = 3225140, hasCoordinate=True, limit=0)['count']]
outTotal
```

```python
#count taxon in Europe
outeurope = occ.search(continent='europe', taxonKey = 3225140, limit=0)['count'] 
outeurope
```

```python
#refine taxon to Portugal
outPT = occ.search(country='PT', taxonKey = 3225140, limit=0)['count'] 
outPT
```

```python
#dictionary of taxon in Europe
Myct= occ.search(continent='europe', taxonKey=3225140, hasCoordinate=True, limit=1000)
print(Myct)
```

```python
#details in data frame for 'results' 
Myctdf = pd.DataFrame.from_dict(Myct['results'])

Myctdf.head(15)
```

```python
#list of last 10 in data frame
df = pd.DataFrame()
Myctdf2 = pd.concat([df, pd.DataFrame.from_dict(Myct['results'])], axis = 0, ignore_index = True, sort = True)
print(Myctdf2.shape)
Myctdf2.tail(10)
```

```python
#import maps from folium
import math
import folium
from folium import Choropleth, Circle, Marker
from folium.plugins import HeatMap, MarkerCluster
```

```python
# import map with european coordinates
f = folium.Figure(width=1000, height=500)
m_2 = folium.Map(location=[54.525961, 15.255119], tiles='openstreetmap', zoom_start=5).add_to(f)
m_2
```

```python
#taxon dot distribution in Europe
for idx, row in Myctdf.iterrows():
  Marker([row['decimalLatitude'], row['decimalLongitude']],popup=Myctdf.iloc[idx]['species']).add_to(m_2)

#display the map
m_2
```

```python
#import Portugal coordinates
f = folium.Figure(width=1000, height=500)
m_3 = folium.Map(location=[38.7, -9.12], tiles='openstreetmap', zoom_start=5).add_to(f)
m_3
```

```python

#taxon dot distribution in Portugal
for idx, row in Myctdf.iterrows():
  Marker([row['decimalLatitude'], row['decimalLongitude']],popup=Myctdf.iloc[idx]['species']).add_to(m_3)

#display the map
m_3
```
